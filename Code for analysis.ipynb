```python
# Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, bartlett, shapiro, anderson, kstest, boxcox
from statsmodels.multivariate.manova import MANOVA
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.utils import resample
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import shapiro, anderson, kstest
from sklearn.decomposition import PCA


def load_data():
    data1 = pd.read_csv('data1.csv', parse_dates=['Date'])  # Parse 'Date' column as datetime, adjust if necessary
    data2 = pd.read_csv('data2.csv', parse_dates=['Date'])  # Adjust date column as needed
    return data1, data2

# Step 1: Handle missing data
def handle_missing_data(df):
    # Check if there are missing values and fill them with the mean of each column
    df = df.fillna(df.mean())
    return df
# Function to apply PCA on highly correlated variables
def apply_pca_on_price_data(df):
    price_columns = ['Open', 'Close', 'High', 'Low']
    if all(column in df.columns for column in price_columns):
        # Standardize and apply PCA
        scaler = StandardScaler()
        price_data_scaled = scaler.fit_transform(df[price_columns])
        pca = PCA(n_components=2)  # Choose number of components to capture desired variance
        price_pca = pca.fit_transform(price_data_scaled)

        # Add the PCA components back to the DataFrame
        df['PC1_Price'] = price_pca[:, 0]
        df['PC2_Price'] = price_pca[:, 1]

        # Optionally drop the original price columns to avoid redundancy
        df.drop(columns=price_columns, inplace=True)

        print(f"Explained Variance by PCA components: {pca.explained_variance_ratio_}")

    return df
# Step 2: Normalize the Data (exclude date/time columns)
def normalize_data(df):
    # Select only numeric columns (excluding 'Date') and fill missing values with the column mean
    numerical_data = df.select_dtypes(include=[np.number]).fillna(df.mean())

    # Standardize the data (mean=0, variance=1)
    scaler = StandardScaler()
    normalized_data = pd.DataFrame(scaler.fit_transform(numerical_data), columns=numerical_data.columns)

    return normalized_data

# Step 3: Initial Exploratory Data Analysis (EDA)
def exploratory_data_analysis(df, dataset_name):
    print(f"\n--- EDA for {dataset_name} ---\n")
    print(df.describe())  # Summary statistics
    print(f"\nMissing values in {dataset_name}: \n", df.isnull().sum())  # Missing values

    # Exclude 'Date' from histograms and numeric operations
    df_numeric = df.select_dtypes(include=[np.number])

    # Visualizations
    df_numeric.hist(figsize=(10, 8))  # Histogram of numeric columns
    plt.suptitle(f"Histograms of {dataset_name}")
    plt.show()
# Heatmap of Correlation Matrix
def plot_correlation_heatmap(df, dataset_name):
    plt.figure(figsize=(10, 8))
    correlation_matrix = df.select_dtypes(include=[np.number]).corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title(f"Correlation Heatmap - {dataset_name}")
    plt.show()

def combine_mfi_rsi(df):
    # Check if both MFI and RSI columns exist to avoid errors
    if 'MFI_14' in df.columns and 'RSI_14' in df.columns:
        # Create the new combined feature as the average of MFI and RSI
        df['Combined_MFI_RSI'] = (df['MFI_14'] + df['RSI_14']) / 2
        # Optionally, drop the original MFI and RSI columns to reduce redundancy
        df.drop(columns=['MFI_14', 'RSI_14'], inplace=True)
        print("Combined MFI and RSI into a new feature 'Combined_MFI_RSI'")
    return df

def plot_trends_separately(df, dataset_name):
    if 'Date' in df.columns:
        df.set_index('Date', inplace=True)

    columns_to_plot = df.columns
    fig, axes = plt.subplots(len(columns_to_plot), 1, figsize=(10, 12), sharex=True)
    fig.suptitle(f"Trends over Time in {dataset_name}")

    for i, column in enumerate(columns_to_plot):
        axes[i].plot(df.index, df[column], label=column, color='tab:blue')  # Continuous line plot

        axes[i].set_ylabel(column)
        axes[i].legend(loc='upper left')

    plt.xlabel('Date')
    plt.tight_layout(rect=[0, 0, 1, 0.98])  # Adjust rect if title is overlapping with graphs
    plt.show()


# Step 4: Fit a Multivariate Normal Distribution and Detect Outliers
def fit_multivariate_normal(df, dataset_name):
    print(f"\n--- Fitting Multivariate Normal for {dataset_name} ---\n")

    # Exclude the 'Date' column if it exists
    df_numeric = df.select_dtypes(include=[np.number])  # Select only numeric columns

    # Fit multivariate normal
    mean_vector = np.mean(df_numeric, axis=0)
    cov_matrix = np.cov(df_numeric.T)

    # Calculate Mahalanobis distance for each observation
    distances = np.diag(np.dot(np.dot((df_numeric - mean_vector), np.linalg.inv(cov_matrix)),
                              (df_numeric - mean_vector).T))

    # Outliers: Those with a high Mahalanobis distance
    threshold = np.percentile(distances, 95)
    outliers = df[distances > threshold]
    outliers_d = df.index[distances > threshold]
    print(f"Outliers in {dataset_name}:")
    print(outliers)
    return mean_vector, cov_matrix, outliers_d


# Step 5: Covariance and Mean Comparison
# Covariance and Mean Comparison with PC1_Price and PC2_Price
def compare_covariance_and_mean(df1, df2):
    print("\n--- Covariance and Mean Comparison ---\n")

    # Updated list of columns to compare, now using Combined_MFI_RSI
    columns_to_compare = ['PC1_Price', 'PC2_Price', 'Volume', 'VWAP', 'Combined_MFI_RSI']

    for column in columns_to_compare:
        print(f"\nComparing '{column}' column:\n")

        # Bartlett’s test to compare covariances
        _, p_value_cov = bartlett(df1[column], df2[column])
        print(f"p-value for covariance comparison (column {column}): {p_value_cov}")

        # T-test for means comparison
        t_stat, p_value_mean = ttest_ind(df1[column], df2[column])
        print(f"p-value for mean comparison (column {column}): {p_value_mean}")

# PCA Analysis Function with Scatter and Explained Variance Plots
def pca_analysis(df, dataset_name):
    print(f"\n--- PCA for {dataset_name} ---\n")

    # Check if 'Date' column exists before dropping it
    if 'Date' in df.columns:
        df_no_date = df.drop(columns=['Date'])
    else:
        df_no_date = df  # If 'Date' column is not present, use df as is

    # Standardize the data
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df_no_date)

    # Perform PCA
    pca = PCA()
    pca_data = pca.fit_transform(scaled_data)

    # Explained variance ratio and cumulative variance
    explained_variance = pca.explained_variance_ratio_
    cumulative_variance = np.cumsum(explained_variance)

    print("Explained Variance Ratio by Principal Components:", explained_variance)
    print("Cumulative Variance Explained:", cumulative_variance)

    # Select the first N components that explain at least 95% of the variance
    n_components = np.argmax(cumulative_variance >= 0.99) + 1
    if n_components < 2:
        n_components = 2  # Ensure we select at least 2 components for visualization

    print(f"Selected {n_components} Principal Components")

    # Create DataFrame for selected PCA components
    pca_df = pd.DataFrame(pca_data[:, :n_components], columns=[f'PC{i+1}' for i in range(n_components)])

    # Add back the Date column if it exists
    if 'Date' in df.columns:
        pca_df['Date'] = df['Date'].values

    # Save the result to CSV
    pca_df.to_csv(f'pca_{dataset_name}.csv', index=False)

    # Plot cumulative variance explained for all components
    plt.figure(figsize=(8, 5))
    plt.plot(cumulative_variance, marker='o')
    plt.title(f'Cumulative Variance Explained by PCA Components - {dataset_name}')
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Variance Explained')
    plt.show()

    # Scatter Plot of the first two principal components
    plt.figure(figsize=(8, 5))
    plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.6, edgecolor='k')
    plt.title(f'PCA Scatter Plot - {dataset_name}')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.grid(True)
    plt.show()

    # Explained Variance Plot for the first two components
    plt.figure(figsize=(6, 4))
    plt.bar(['PC1', 'PC2'], explained_variance[:2], alpha=0.7, edgecolor='k')
    plt.title(f'Explained Variance by Principal Components - {dataset_name}')
    plt.ylabel('Variance Explained')
    plt.show()

    return pca_df


# Check normality with transformations
def check_normality_with_transformations(df, dataset_name):
    print(f"\n--- Checking Normality for {dataset_name} ---\n")
    normality_results = {"Original": {}, "Log": {}, "Square Root": {}, "Box-Cox": {}}
    numeric_columns = df.select_dtypes(include=[np.number]).columns

    def normality_tests(column_data):
        shapiro_stat, shapiro_p = shapiro(column_data)
        ks_stat, ks_p = kstest(column_data, 'norm')
        anderson_stat = anderson(column_data).statistic
        anderson_critical = anderson(column_data).critical_values
        return {"Shapiro p-value": shapiro_p, "KS p-value": ks_p, "Anderson stat": anderson_stat, "Anderson critical": anderson_critical}

    for column in numeric_columns:
        normality_results["Original"][column] = normality_tests(df[column])

        if (df[column] > 0).all():
            normality_results["Log"][column] = normality_tests(np.log(df[column] + 1))
            normality_results["Box-Cox"][column] = normality_tests(boxcox(df[column] + 1)[0])
        else:
            normality_results["Log"][column] = None
            normality_results["Box-Cox"][column] = None

        normality_results["Square Root"][column] = normality_tests(np.sqrt(df[column] + 1))

    passes_normality = {"Original": True, "Log": True, "Square Root": True, "Box-Cox": True}

    for transform in normality_results.keys():
        for column, result in normality_results[transform].items():
            if result and result["Shapiro p-value"] < 0.05:
                passes_normality[transform] = False

    for transform, passes in passes_normality.items():
        if passes:
            print(f"{dataset_name} passed normality test with {transform} transformation.")
            return transform
    print(f"{dataset_name} did not pass normality test with any transformation.")
    return None

# Conditional Profile Analysis
# Profile Analysis with updated columns
def profile_analysis(df1, df2):
    print("\n--- Profile Analysis ---\n")
    df1['Dataset'] = 'Dataset 1'
    df2['Dataset'] = 'Dataset 2'
    combined_data = pd.concat([df1, df2], ignore_index=True)
    combined_data['Dataset'] = ['Dataset 1'] * len(df1) + ['Dataset 2'] * len(df2)

    # Updated variables for profile analysis, now using Combined_MFI_RSI
    combined_data_melted = pd.melt(combined_data, id_vars=['Dataset'], value_vars=['PC1_Price', 'PC2_Price', 'Volume', 'VWAP', 'Combined_MFI_RSI'])
    combined_data_melted.columns = ['Dataset', 'Variable', 'Value']

    # Parallelism test
    formula = 'Value ~ C(Dataset) * C(Variable)'
    model = ols(formula, data=combined_data_melted).fit()
    anova_results = anova_lm(model, typ=2)
    print("\nParallelism Test (Interaction between Dataset and Variable):")
    print(anova_results)

    # Level Equality test
    formula = 'Value ~ C(Dataset)'
    model = ols(formula, data=combined_data_melted).fit()
    anova_results = anova_lm(model, typ=2)
    print("\nLevel Equality Test (Main effect of Dataset):")
    print(anova_results)

    # Flatness test
    formula = 'Value ~ C(Variable)'
    model = ols(formula, data=combined_data_melted).fit()
    anova_results = anova_lm(model, typ=2)
    print("\nFlatness Test (Main effect of Variable):")
    print(anova_results)

# Conditional MANOVA
# MANOVA Analysis with updated columns
def manova_analysis(df1, df2):
    print("\n--- MANOVA Analysis ---\n")
    df1['Dataset'] = 'Dataset 1'
    df2['Dataset'] = 'Dataset 2'
    combined_data = pd.concat([df1, df2], ignore_index=True)
    combined_data['Dataset'] = ['Dataset 1'] * len(df1) + ['Dataset 2'] * len(df2)

    # Updated MANOVA formula with Combined_MFI_RSI
    manova = MANOVA.from_formula('PC1_Price + PC2_Price + Volume + VWAP + Combined_MFI_RSI ~ Dataset', data=combined_data)
    manova_results = manova.mv_test()
    print(manova_results)

    # Interpretation
    wilks_lambda_test = manova_results.results['Dataset']['stat']
    p_value = wilks_lambda_test.loc["Wilks' lambda", 'Pr > F']
    if p_value < 0.05:
        print("Interpretation: Statistically significant difference in the variables across datasets (Wilks' Lambda test).")
    else:
        print("Interpretation: No statistically significant difference across datasets (Wilks' Lambda test).")

def custom_permanova(distance_matrix, grouping, permutations=999):
    unique_groups = np.unique(grouping)
    group_means = {group: np.mean(distance_matrix[grouping == group, :][:, grouping == group]) for group in unique_groups}
    overall_mean = np.mean(distance_matrix)
    ss_between = sum([(np.sum(grouping == group) * (group_means[group] - overall_mean) ** 2) for group in unique_groups])

    permuted_ss_between = []
    for _ in range(permutations):
        permuted_grouping = resample(grouping)
        permuted_group_means = {group: np.mean(distance_matrix[permuted_grouping == group, :][:, permuted_grouping == group]) for group in unique_groups}
        permuted_ss_between.append(sum([(np.sum(permuted_grouping == group) * (permuted_group_means[group] - overall_mean) ** 2) for group in unique_groups]))

    p_value = np.mean(permuted_ss_between >= ss_between)
    return {'ss_between': ss_between, 'p-value': p_value, 'permutations': permutations}
# Clustering and PERMANOVA if Normality Fails
def clustering_and_permanova_analysis(df1, df2, n_clusters=3):
    combined_df = pd.concat([df1, df2], ignore_index=True)
    combined_df['Dataset'] = ['Dataset 1'] * len(df1) + ['Dataset 2'] * len(df2)
    numeric_columns = combined_df.select_dtypes(include=[np.number]).columns

    scaler = StandardScaler()
    normalized_data = scaler.fit_transform(combined_df[numeric_columns])

    # KMeans Clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    combined_df['Cluster'] = kmeans.fit_predict(normalized_data)
    silhouette_avg = silhouette_score(normalized_data, combined_df['Cluster'])
    print(f"\nSilhouette Score for Clustering: {silhouette_avg}")

    # Visualize Clusters with PCA
    pca = PCA(n_components=2)
    pca_data = pca.fit_transform(normalized_data)
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=combined_df['Cluster'], palette='viridis', s=60)
    plt.title("PCA Visualization of Clusters")
    plt.xlabel("Principal Component 1")
    plt.ylabel("Principal Component 2")
    plt.show()

    # PERMANOVA on Clusters
    distance_matrix = squareform(pdist(normalized_data, metric='euclidean'))
    permanova_results = custom_permanova(distance_matrix, combined_df['Cluster'])
    p_value = permanova_results['p-value']
    print(f"\nPERMANOVA Results:\n{permanova_results}")

    # Interpretation based on PERMANOVA p-value
    if p_value < 0.05:
        print("Interpretation: The PERMANOVA results indicate a significant difference between clusters (p-value < 0.05), suggesting that each cluster represents a distinct group in multivariate space.")
    else:
        print("Interpretation: The PERMANOVA results do not indicate a significant difference between clusters (p-value ≥ 0.05), implying that the clustering may not strongly represent separate groups.")
# Visualize Outliers with Pair Plot
# Visualize Outliers with Pair Plot
def plot_outliers_pairplot(df, outlier_indices, dataset_name):
    # Create a copy of numeric data and reset index
    df_numeric = df.select_dtypes(include=[np.number]).copy().reset_index(drop=True)

    # Create 'Outlier' label column for each row
    df_numeric['Outlier'] = 'Normal'

    # Map outlier indices to the re-indexed DataFrame
    outlier_indices = [i for i in outlier_indices if i < len(df_numeric)]
    df_numeric.loc[outlier_indices, 'Outlier'] = 'Outlier'

    # Pair Plot with outliers highlighted
    plt.figure(figsize=(12, 6))
    sns.pairplot(df_numeric, hue='Outlier', palette={'Normal': 'blue', 'Outlier': 'red'}, diag_kind="kde", markers=["o", "s"])
    plt.suptitle(f"Pair Plot of Variables with Outliers - {dataset_name}", y=1.02)
    plt.show()

# Main analysis function
def run_analysis():
    data1, data2 = load_data()
    data1 = handle_missing_data(data1)
    data2 = handle_missing_data(data2)



    # Combine MFI and RSI into a single feature for both datasets

    exploratory_data_analysis(data1, 'Dataset 1')
    exploratory_data_analysis(data2, 'Dataset 2')

    plot_trends_separately(data1, 'Dataset 1')
    plot_trends_separately(data2, 'Dataset 2')

    plot_correlation_heatmap(data1, 'Dataset 1')
    plot_correlation_heatmap(data2, 'Dataset 2')

    data1 = apply_pca_on_price_data(data1)
    data2 = apply_pca_on_price_data(data2)

    data1 = combine_mfi_rsi(data1)
    data2 = combine_mfi_rsi(data2)

    # Apply PCA on highly correlated price data at the start of the analysis


    plot_correlation_heatmap(data1, 'Dataset 1')
    plot_correlation_heatmap(data2, 'Dataset 2')

    normalized_data1 = normalize_data(data1)
    normalized_data2 = normalize_data(data2)



    mean1, cov1, out1 = fit_multivariate_normal(normalized_data1, 'Dataset 1')
    mean2, cov2, out2 = fit_multivariate_normal(normalized_data2, 'Dataset 2')

    plot_outliers_pairplot(data1, out1, 'Dataset 1')
    plot_outliers_pairplot(data2, out2, 'Dataset 2')

    # Drop outliers using iloc to ensure removal by position
    data1 = data1.drop(data1.index[list(out1)])
    data2 = data2.drop(data2.index[list(out2)])
    print("Completed outliers removal. New shape of data:")
    print(data1.shape, data2.shape)

    # Covariance and Mean Comparison
    compare_covariance_and_mean(data1, data2)

    normality_check1 = check_normality_with_transformations(data1, 'Dataset 1')
    normality_check2 = check_normality_with_transformations(data2, 'Dataset 2')

    if normality_check1 and normality_check2:
        print("Proceeding with Profile Analysis and MANOVA as normality check passed.")
        profile_analysis(data1, data2)
        manova_analysis(data1, data2)
    else:
        print("Normality check failed, proceeding with Clustering and PERMANOVA as an alternative.")
        clustering_and_permanova_analysis(data1, data2)

    # Step 8: PCA
    pca_analysis(normalized_data1, 'Dataset 1')
    pca_analysis(normalized_data2, 'Dataset 2')


    # Run the full analysis when the script is executed
if __name__ == "__main__":
    run_analysis()
```
